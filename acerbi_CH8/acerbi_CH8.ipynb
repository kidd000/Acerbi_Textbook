これまでの章では、文化の進化のダイナミクス、つまり、異なるバイアスや伝達経路が、時間の経過とともに集団内の文化的特徴の頻度にどのように影響を与えるかについて述べてきました。しかし、同じように重要なのは、一歩下がって、文化が存在するという可能性がそもそもどこから来たのかを考えることです。つまり、文化の進化、そして文化進化の進化についても考える必要があります。

ここで私たちが問うことのできる最も基本的な疑問は、なぜ個人学習（環境から直接、自分自身で学ぶ）に比べて、社会的な学習（他者から学ぶ）の能力が進化したのかということです。この質問への直感的な答えは、社会的な学習は、個人学習よりもコストがかからないということです。毒かもしれない食品を試食してみることを想像してみてください。一つの方法は、それぞれ試食して気分が悪くなるか確認することです。より危険の少ない戦略は、自分の隣人を観察し、彼らが食べているものを食べることであろう。彼らがいつも病弱に見えない限り、これはおそらくおいしい（そして進化的に適応的な）選択につながるだろう。その結果、社会学習は集団の平均適応度を高めるはずである。

しかし、この直感は誤解を招く可能性がある。これは1988年にAlan Rogersによって今では古典的な社会的学習の進化のモデルで示された(Rogers (1988))。このモデルはしばしば「ロジャースのパラドックス」と呼ばれているが、これは、ある条件下では、社会学習が個人学習よりもコストがかからない場合でも、適応度の増加にはつながらないことを示しているからである。より正確には、社会的学習者を含む集団の平均適合度は、個人学習者のみで構成される集団の平均適合度を超えない。ここでは、ロジャースの数学モデルを個人ベースのシミュレーションで再現し、いつ、なぜこのような不可解な結果が生じるのかを確認する。

# **8.1　　ロジャースのパラドクスをモデル化する**

ロジャースのモデルでは、N人の個人が存在する。各個人は固定の学習戦略を持っている：個人学習者か社会学習者である。各個人はまた、前章で述べた特徴を整数（例えば "5 "や "32"）で表す行動を示す。(「形質」と「行動」は文化進化の文献ではしばしば互換的に使われている)。環境状態Eもあり、これも整数で表される。個人の行動が環境と一致した場合、一致しない場合と比較して、彼らは適応度が増加する。一致している場合は「おいしい食べ物」を表し、不一致の場合は「有毒な食べ物」を表すと考えても良い。

各世代において、個々の学習者は環境を直接サンプリングし、環境にマッチした「正しい」適応行動を獲得する確率pを持つ（したがって、正しくない、不適応行動を採用する確率1-pを持つ）。社会学習者は前世代のメンバーをランダムに選び、第1章で検討したバイアスのない伝達の場合と同様に、その行動をコピーする。

以前のモデルとは異なり、ここでは行動や形質ではなく、学習戦略が時間の経過とともにどのように進化していくかに関心がある。したがって、我々は集団における社会学習者の割合を追跡したいと考えており、これを我々はpSLと呼ぶ（1-pSLは個人学習者の割合である）。これらの学習戦略は、親から子孫へ（おそらく遺伝的に、あるいは文化的に）継承され、その戦略の担い手の適応度の影響を受けると仮定する。そのため、適応度パラメータを指定する必要があります。

各個体はベースライン・フィットネスwからスタートします。
これは通常１に設定されているが、これはトリッキーで扱いにくい負の適応度になることを避けるためである。環境にマッチした行動をとる個体は、適応度が+b増加する。環境に合わない行動をした個人は、適応度のペナルティとして-bを受ける。上記の概要に明示されているのは、社会学習は個人学習よりもコストがかからないということである。したがって、個人学習者の適応度上のコストは-b∗c、社会学習者の適応度上のコストは-b∗sとなり、c>sとなる。単純にするために、s=0（社会的学習はコストフリー）、c>0 とすればよいので、パラメータを 1 つ変更すればよいことになる。

そして、上記をもとに各個体の適応度を集計し、次の世代を作ります。各個体は、他の戦略と比較して、自分の戦略の適応度に比例して繁殖します。
また、繁殖中に何らかの突然変異が起こることも想定しています。確率μで、新しい個体は他の学習戦略に "変異 "する。我々はここでは、社会学習が個人学習からどのように進化するかに興味があるので、我々は完全に個人学習者で構成された第一世代から開始します。その後、社会学習は第二世代以降、突然変異を経て現れる。

最後に、ロジャースは環境変化の影響に興味を持っていました。世代ごとに、環境が新しい状態に変化する確率 u がある。ロジャースのオリジナルモデルでは、環境は同じ2つの状態を行ったり来たりしながら反転していた。しかし、これは環境の変化が非常に速い場合に問題となる。なぜなら、古い行動をとる個体は、環境が以前の状態に反転した場合に適応度の恩恵を受けることができるからである。そこで我々は、環境が変化すると、それまでどの個体も経験したことのない新しい値に変化すると仮定します。

これは複雑なモデルですが、一歩一歩進めてみましょう。まず、前の章と同じように、出力と個体のpopulationを格納するための データフレーム を作成して初期化します。ここでの出力は、これまでと同様に、rmax（ラン数）とtmax（世代数）のデータを保存するのに十分な大きさにする必要があります。そして、pSL（社会学習者の割合）とW（平均集団適合度）のNAプレースホルダを作成する必要があります。母集団データフレームは、学習戦略（「個人的」または「社会的」）、行動（最初はすべてNA）、適応度（最初はすべてNA）という個人の特性を格納します。最後に、環境Eをゼロで初期化し、その後、環境が変化することを意味する数値の増加を行います。

```
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

N = 100
r_max = 1
t_max = 10

"""アウトプット用のータフレームを作る"""
# run列となるリスト
run = np.repeat(range(1, r_max + 1), repeats=t_max, axis=0)
# generation列となるリスト
generation = np.array(list(range(1, t_max + 1)) * r_max)
# p_SL(社会学習者の頻度)列となるリスト
p_SL = np.zeros(t_max * r_max)
# W(平均集団適応度)列となるリスト
W = np.zeros(t_max * r_max)

output = pd.DataFrame()
output["run"] = run
output["generation"] = generation
output["p_SL"] = p_SL
output["W"] = W

# 毎回このpandasデータフレームに値を代入していたら処理が重くなるので、numpy2次元アレーで代用
p_SL_array = np.zeros((r_max, t_max))
W_array = np.zeros((r_max, t_max))

"""populationデータフレーム(numpy配列)を作る"""
population_learning = np.repeat("individual", repeats=N)
population_behaviour = np.zeros(N)
population_fitness = np.zeros(N)


# 環境状態を0に初期化
E = 0
```

6.  環境変化

以下のコードは個人学習を実装しています。これは第一世代にも適用されます。まず、環境に正しくマッチする個人学習の確率である p に依存する TRUE と FALSE のベクトルを作成します。この確率で、個々の学習者の行動は正しいE値になるよう設定されている。そうでなければ、学習者には正しくない行動E-1が与えられる。

```
'''今はまだ、正常に動作しません'''
# 個人学習者が正答を導けたかどうか、個人学習の成功確率p(環境情報の正確性)に従って、numpyアレーを生成
learn_correct = np.random.choice([True, False], size=N, p=[p, 1-p], replace=True)
# 個人学習を実行
population_behaviour[(learn_correct==True) & (population_learning=="individual")] = E
population_behaviour[(learn_correct==False) & (population_learning=="individual")] = E-1
```

ここで各個人の適応度を求めます。次に、個人の行動が正しいか正しくないかに基づいて、bを足したり引いたりします。最後にコストを課すが、これは社会学習者と個人学習者で異なります。

```
'''今はまだ、正常に動作しません'''
# ベースラインの適応度をすべてのエージェントに割り当てる
population_fitness = np.repeat(w, repeats=N)

# 環境と行動が合致している(適応的な行動をとっている)個人の適応度に b を加える
population_fitness[population_behaviour == E] += np.repeat(b, repeats=sum(population_behaviour==E))
# 環境と行動が合致していない(非適応的な行動をとっている)個人の適応度から b を引く
population_fitness[population_behaviour != E] -= np.repeat(b, repeats=sum(population_behaviour!=E))

# 個人学習者にコスト b*c をかける
population_fitness[population_learning == "individual"] -= np.repeat(b*c, repeats=sum(population_learning == "individual"))
# 社会学習者にコスト b*s(これは、0) をかける
population_fitness[population_learning == "social"] -= np.repeat(b*s, repeats=sum(population_learning == "social"))
```

第五段階は繁殖です。ここでは、以前に行ったように、現在のpopulationデータフレームを前のデータフレーム（previous_population）と呼ばれる新しいデータフレームに入れます。これは、適応度を計算するための記録としての役割と、上で説明した社会学習段階のためのデモンストレータのソースとしての役割の両方を果たします。これを行った後、現在の母集団の行動と適応度をリセットします。次に、適応度に基づいて学習戦略を上書きします。

まず、集団全体の適応度に対する個々の学習者の適応度である fitness_IL を取得します（個人学習者がいると仮定します。これは、次の世代で新しい個人を個人学習者として設定する確率として機能します。我々は関数sample()を使用して、学習戦略を変異させる個体の確率を表す突然変異リストを作成します。最後に、「突然変異」した個体の学習戦略を変更します。最初の突然変異命令で個人学習から社会学習に変更された個体を2回も突然変異させないようにするために、一時的に新しいデータフレーム、previous_poulation2を作成する必要があることに注意してください。

最終段階が最も簡単です。確率uで環境状態Eを1つ増やす。それ以外の場合は、そのままの状態を維持します。これを行うには、0から1の間の乱数を選び、uがこれを超えるとEの値を増やします。

```
'''今はまだ、正常に動作しません'''
if np.random.uniform(low=0,high=1,size=1) < u:
    E += 1
```

これで各世代で発生する6つのステージをカバーしています。これらをすべて、ループするランとループする世代にまとめることができるようになりました。また、これらすべてを関数の中に入れることもできます。これは以前の章でお馴染みのことでしょう。ほとんどすべてのコードは上から抜粋したもので、異なるステージに番号を付けています（上記のチャンクの中の特定のコード行のコメントを見つけることができます）。また、負の適応度を取得しないことを確認するために、最初にパラメータチェックを追加しています。もう一つ新しいポイントは、関数呼び出しの際にいくつかのパラメータ（w, b, s）をすでに設定していることです。この方法では、関数が呼び出されたときに指定されていなければ、パラメータはこれらのデフォルト値に設定されます。他のパラメータは代わりに指定する必要があります。

```
def rogers_model(N, t_max, r_max, c, mu, p, u, w = 1.0, b = 0.5, s = 0):
    # マイナスの適応度を生むようなパラメタ設定にしていないかチェックを行う
    if (b*(1+c) > 1) or (b*(1+s) > 1):
        return print("無効なパラメタ設定だよ！")
    
    """アウトプット用のータフレームを作る"""
    # run列となるリスト
    run = np.repeat(range(1, r_max + 1), repeats=t_max, axis=0)
    # generation列となるリスト
    generation = np.array(list(range(1, t_max + 1)) * r_max)
    # p_SL(社会学習者の頻度)列となるリスト
    p_SL = np.zeros(t_max * r_max)
    # W(平均集団適応度)列となるリスト
    W = np.zeros(t_max * r_max)

    output = pd.DataFrame()
    output["run"] = run
    output["generation"] = generation
    output["p_SL"] = p_SL
    output["W"] = W

    # 毎回このpandasデータフレームに値を代入していたら処理が重くなるので、numpy2次元アレーで代用
    p_SL_array = np.zeros((r_max, t_max))
    W_array = np.zeros((r_max, t_max))

    for r in range(r_max):
        """populationデータフレーム(numpy配列)を作る"""
        population_learning = np.repeat("individual", repeats=N)
        population_behaviour = np.zeros(N)
        population_fitness = np.zeros(N)

        # 環境状態を0に初期化
        E = 0

        for t in range(t_max):
            '''1. 社会学習フェイズ'''
            if sum(population_learning == "social") > 0:
                population_behaviour[population_learning=="social"] = np.random.choice(previous_population_behaviour,
                                                                      size=sum(population_learning=="social"),
                                                                      replace = True)
            '''2. 個人学習フェイズ'''
            # 個人学習者が正答を導けたかどうか、個人学習の成功確率p(環境情報の正確性)に従って、numpyアレーを生成
            learn_correct = np.random.choice([True, False], size=N, p=[p, 1-p], replace=True)
            # 個人学習を実行
            population_behaviour[(learn_correct==True) & (population_learning=="individual")] = E
            population_behaviour[(learn_correct==False) & (population_learning=="individual")] = E-1

            '''3. 適応度算出フェイズ'''
            # ベースラインの適応度をすべてのエージェントに割り当てる
            population_fitness = np.repeat(w, repeats=N)

            # 環境と行動が合致している(適応的な行動をとっている)個人の適応度に b を加える
            population_fitness[population_behaviour == E] += np.repeat(b, repeats = sum(population_behaviour == E))
            # 環境と行動が合致していない(非適応的な行動をとっている)個人の適応度から b を引く
            population_fitness[population_behaviour != E] -= np.repeat(b, repeats = sum(population_behaviour != E))
 
            # 個人学習者にコスト b*c をかける
            population_fitness[population_learning == "individual"] -= np.repeat(b*c, repeats=sum(population_learning == "individual"))
            # 社会学習者にコスト b*s(これは、0) をかける
            population_fitness[population_learning == "social"] -= np.repeat(b*s, repeats=sum(population_learning == "social"))

            '''4. アウトプット用のデータフレームに格納'''
            # 社会学習者のpopulationにおける割合をnumpy2次元アレーに格納
            p_SL_array[r,t] = sum(population_learning == "social") / N
            # populationの平均適応度をnumpy2次元アレーに格納
            W_array[r,t] = np.average(population_fitness)

            '''5. 繁殖(複製)フェイズ'''
            # 前の世代の行動と適応度として格納
            previous_population_learning = population_learning
            previous_population_behaviour = population_behaviour
            previous_population_fitness = population_fitness
            # 現世代の行動と適応度をリセット
            population_behaviour = np.zeros(N)
            population_fitness = np.zeros(N)

            # 個人学習者がいた場合、相対適応度 fitness_IL を算出する
            if sum(previous_population_learning == "individual") > 0:
                fitness_IL = sum(previous_population_fitness[previous_population_learning == "individual"])/sum(previous_population_fitness)
            else:
                fitness_IL = 0

            # この相対適応度をもとに、新しいpopulationを作る
            population_learning = np.random.choice(["individual", "social"], size=N,p=[fitness_IL, 1-fitness_IL], replace=True)

            # 突然変異が起きるかどうか、確率muに従って、リストを作る
            mutation = np.random.choice([True, False], size=N, p=[mu, 1-mu], replace=True)
            # 2回突然変異が起きるのを防ぐために別のリストにコピー
            previous_population_learning_2 = previous_population_learning

            # 突然変異の対象が個人学習者だった場合、社会学習者に転向
            population_learning[(mutation==True) & (previous_population_learning_2=="individual")] = "social"
            # 突然変異の対象が社会学習者だった場合、個人学習者に転向
            population_learning[(mutation==True) & (previous_population_learning_2=="social")] = "individual"

            '''6. 環境変化フェイズ'''
            if np.random.uniform(low=0,high=1,size=1) < u:
                E += 1
    # numpy2次元アレーをpandasのデータフレームへとラン毎に代入
    output = output.assign(p_SL = p_SL_array.flatten(), W = W_array.flatten())

    #結果（プロットするデータ）をリターン
    return output
```

```
# 結果をプロットするための関数を定義
import seaborn as sns

'''社会学習者の頻度をプロットする関数'''
def plot_multiple_runs_p_SL(data_model, t_max):
  # 体裁いじいじ
  sns.set_palette("hls",n_colors=24)
  sns.set('poster', 'whitegrid', 'dark', font_scale=1,
          rc={"lines.linewidth": 2, 'grid.linestyle': '--'})
  
  # ランごとの平均値推移
  fig, ax = plt.subplots(figsize=(12, 8))
  sns.lineplot(x="generation", y="p_SL", hue="run", data=data_model, legend = "full", palette="hls")

  # 全ランの平均値推移
  p_mean_list = []
  for g in range(1, t_max+1):
    p_mean_list.append(data_model[data_model.generation == g].p_SL.mean())

  plt.plot(p_mean_list,marker="None",linestyle="-",linewidth=3.0,color="black")
  plt.legend(bbox_to_anchor=(1.01, 1),borderaxespad=0, title="run")
  plt.ylim(-0.1, 1.1)
  plt.xlabel("Generation")
  plt.ylabel("proportion of social learners")
  plt.show()

'''集団の平均適応度をプロットする関数'''
def plot_W(data_model, t_max, c, p, w = 1, b = 0.5):
    # 体裁いじいじ
  sns.set_palette("hls",n_colors=24)
  sns.set('poster', 'whitegrid', 'dark', font_scale=1,
          rc={"lines.linewidth": 2, 'grid.linestyle': '--'}) 

  # ランごとの平均値推移
  fig, ax = plt.subplots(figsize=(12, 8))
  sns.lineplot(x="generation", y="W", hue="run", data=data_model, legend = "full", palette="hls", zorder=1)

  # 全ランの平均値推移
  p_mean_list = []
  for g in range(1, t_max+1):
    p_mean_list.append(data_model[data_model.generation == g].W.mean())
  plt.plot(p_mean_list,marker="None",linestyle="-",linewidth=3.0,color="black",zorder=2)

  # 個人学習者の適応度（点線）    
  plt.hlines(y = w + b * (2 * p - c -1), xmin=1, xmax=t_max, colors="black", linestyles='dashed', zorder=3) 

  plt.legend(bbox_to_anchor=(1.01, 1),borderaxespad=0, title="run")
  plt.ylim(0, None) 
  plt.xlabel("Generation")
  plt.ylabel("mean population fitness")
  plt.show()
```

これで、1000人のpopulationで10ラン、200世代のシミュレーションを実行することができます。設定した他のパラメータは、個人学習に関連するコスト（c = 0.9）、突然変異率、すなわち、学習戦略を継承した個体が他の学習戦略に切り替わる確率（mu = 0.001）、個人学習の精度（p = 1）、そして最後に、環境変化の確率（u = 0.2）です。これらのパラメータの他の値の影響については後で見ますが、自由に変更して、何が起こるかを見てみてください。

```
data_model = rogers_model(N = 1000, t_max = 200, r_max = 10, c = 0.9, mu = 0.01, p = 1, u = 0.2)
```

```
plot_multiple_runs_p_SL(data_model=data_model, t_max = 200)
```

ここでは、これらのパラメータ値では、社会学習者の平均割合はすぐに0.5になりますが、その後はこの値の周辺で変動したままであることがわかります。しかし、各ランは非常に不規則で、大きなばらつきがあります。しかし、ロジャースのパラドックスを理解する上でより重要なのは、母集団の平均適応度であり、これが個々の学習者だけで構成された母集団とどのように比較されるかである。そこで、時間経過に伴う平均的な母集団の適応度をプロットする必要がある。これは、rogers_model()関数の出力のWです。以下の関数は、これを個人学習者の適応度を表す点線とともにプロットします。個人学習者の適応度は固定されており、事前にわかっており、シミュレーションのパラメータの値を知っていれば計算することができます。プロット関数には、いくつかの新しい要素があります。まず、data_model populationと一緒に、シミュレーションのパラメータに関する他の情報を関数に渡して、個人学習者の推移を描くことができるようにします。

```
plot_W(data_model=data_model, t_max=200, c=0.9, p=1.0, w = 1, b = 0.5)
```

これがロジャースのパラドックスである。社会学習の方が個人学習よりもコストが低い（すなわちs<c）にもかかわらず、約50％社会学習者が占める集団は、個人学習者の集団のフィットネスを示す点線を一貫して超えていない。社会学習は適応度を増加させない。このことは、社会学習を中心とした文化が、我々の種の生態学的成功の鍵を握ってきたという一般的な主張にも反している。

この結果の理由は、社会学習は変化する環境下では頻度に依存しているからである。個人学習者はコストのかかる個人学習を受け、正しい行動を発見し、最初はうまくいく。社会的学習者はその後、その行動をコピーするが、より低コストである。したがって、社会学習者は個人学習者よりも優れた行動をとり、個人学習者を凌駕する。しかし、環境が変わると、社会学習者は時代遅れの行動をコピーしたままになってしまうため、うまくいかない。個人学習者は新しい環境の状態を検出できるので、個人学習者の方がうまくいく。個々の学習者は回数を増やし、そのサイクルを続けます。これが一つのランにおける大きな変動が示していることです。分析的には、社会学習者と個人学習者の頻度が同じという均衡に到達することが示されますが、定義上、この均衡は、個々の学習者で完全に構成された集団と同じ平均的な適応度を持たなければなりません。これが「パラドックス」である。

これをさらに探求するために、パラメータを変更することができます。まず、個人学習のコストをc=0.9からc=0.4に下げることができます。

```
data_model_2 = rogers_model(N = 1000, t_max = 200, r_max = 10, c = 0.4, mu = 0.01, p = 1, u = 0.2)
```

```
plot_multiple_runs_p_SL(data_model=data_model_2, t_max = 200)
```

```
plot_W(data_model=data_model_2, t_max = 200, c=0.4, p=1.0)
```

予想通り、この設定の変更は、個人学習者の学習に対するペナルティコストを軽減することになるので、社会学習者の割合を減らすことになる。また、予想されるように、パラドックスも残っている。実際には、より多くの個人学習者がいることを考えると、それはさらに明白である。

また同様に、個人学習の精度pを1から0.7に下げることができる。


```
data_model_3 = rogers_model(N = 1000, t_max = 200, r_max = 10, c = 0.9, mu = 0.01, p = 0.7, u = 0.2)
```

```
plot_multiple_runs_p_SL(data_model = data_model_3, t_max = 200)
```

```
plot_W(data_model=data_model_3, t_max=200, c = 0.9, p = 0.7)
```

今度は社会学習者が大部分を占めます。しかし、パラドックスは依然として残っています：大多数の社会学習者は、純粋な個人学習の適応度ラインを実際には超えていません。

上記の説明が正しいならば、環境を一定にすることでパラドックスは解消されるはずである。環境が同じままであれば、行動は決して時代遅れになることはなく、個人学習者が優位に立つことはないでしょう。u=0を設定すると、このことがわかります。

```
data_model_4 = rogers_model(N = 1000, t_max = 200, r_max = 10, c = 0.9, mu = 0.01, p = 1, u = 0)
```

```
plot_multiple_runs_p_SL(data_model=data_model_4, t_max=200)
```

```
plot_W(data_model=data_model_4, t_max=200, c = 0.9, p = 1)
```

今度はパラドックスは消えました：社会学習者は、個人学習者が正しい行動を発見するのに苦労した後、個人学習者を明らかに上回り、社会学習者は個人学習の適応度を示す点線よりも平均適応度が高くなります。（各ラン内の変動が消えたことにも注目してください）。これは予想通りです。ロジャースのパラドックスは決定的に環境の変化に依存する。しかし、自然が一定の環境を提供することはほとんどありません。食糧源は場所を変え、技術は蓄積され、言語はいくつかの波形に分かれ、気候は変化する。



# **8.2　　モデルのまとめ**

ロジャースのモデルは、明らかに現実を粗く単純化したものです。しかし、前の章で述べたように、現実主義がモデリン グの目的ではないことが多い。モデルは、たとえ単純で著しく非現実的なものであっても、私たちに仮定の中で考え、自然言語で理論化することを強いる。ロジャースのモデルはその良い例である。社会学習が集団の平均的な適応度、つまり適応を高めるべきだというのは理にかなっているように聞こえても、この単純なモデルでは、このような仮定のもとではそうはなりません。一方、社会学習が平均適応度を増加させる状況は、環境が変化しない場合に見られました。しかし、これはあまりもっともらしいとは言えないでしょう。環境は常に変化します。したがって、ロジャースのモデルの他の仮定を検証する必要があります。これは次の章で検討することにしましょう。

# **8.3　　さらに学びたい人のために**

それが学習のコストを減らすので、社会学習が適応的であるという主張の初期の例は、BoydとRicherson(1985)で見つけることができます。Rogers（1988）はその後、この章で見てきたように、この主張に挑戦しました。次の章では、ロジャースのパラドックスに対する「解決策」を検討した、その後のモデルを検討します。
